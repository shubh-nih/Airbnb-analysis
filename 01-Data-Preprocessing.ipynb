## issues with dataset

- The `room type` column mixes different property categories (Entire home/apt).
- `availability 365` column contains negative values and values greater than 365, which are not valid.
- `minimum nights` column contains negative values, which are not logically possible.

### Cleaning

```python
# missing value are less than 1% we can drop those missing value
for col in df.columns:
    df.dropna(subset = [col], inplace = True)
```

```python
df.isnull().sum()
```

```python
# Created a new column apartment. 
df['is_apartment'] = df["room type"].apply(lambda x: True if x == "Entire home/apt" else False)
```

```python
# remove 'apt' from room type
df["room type"] = df["room type"].apply(lambda x: 'Entire home' if x == "Entire home/apt" else x)
```

```python
# remove '$', ',' then convert to float
df["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)
df["service fee"] = df["service fee"].replace('[\$,]', '', regex=True).astype(float)
```

```python
# check
df.sample(5)
```

```python
# remove negative sign from availability 365, minimum nights
df['availability 365'] = df['availability 365'].abs()
df['minimum nights'] = df['minimum nights'].abs()
```

```python
# check
df[df['availability 365'] < 0]
```

```python
# check
df[df['minimum nights'] < 0]
```

```python
# Replaced above 365 to 365.
df['availability 365'] = df['availability 365'].apply(lambda x: 365 if x > 365 else x)
```

### Feature Engineering

```python
# Total stay cost
df['total stay cost'] = (df['price'] + df['service fee']) * df['minimum nights']
```

```python
# Availability category
def available_category(available):
    if available == 0:
        return 'Unavailable'
    elif 1 <= available <= 180:
        return 'Seasonal'
    elif 181 <= available <= 364:
        return 'Mostly available'
    elif available == 365:
        return 'Full time'
```

```python
df['availability category'] = df['availability 365'].apply(available_category)
```
